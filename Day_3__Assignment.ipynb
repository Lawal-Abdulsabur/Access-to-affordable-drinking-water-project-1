1. unit_price needs to be changed to numeric

2. date needs to be chaged to datatime

3. quantity needs to be a numeric variable

4. sevaral columns have missing values

```python
# Before
print("Type before:" , df["unit _ price"].dtype)

print(df["unit _ price"])

```

```python
 # Step- 1 remove the GHS
df["unit _ price"]= df["unit _ price"].str.replace("GHS" , "").str.strip()
# Step-2 Convert to number
df["unit _ price"] = pd.to_numeric(df["unit _ price"], errors="coerce")
print("Type after", df["unit _ price"].dtype)
print(df["unit _ price"].head(3))
```

```python
# Before
print("Type before", df["quantity"].dtype)
print(df["quantity"].head(3))
```

```python
# Before
print("Type before:", df["date"].dtype)
print(df["date"].head(6))
```

```python
df["date"] =  pd.to_datetime(df["date"], format = "mixed")

# After
print("Type after:" , df["date"].dtype)
print(df["date"].head(6))
```

```python
# Extract months and days from the date
df["month"]= df["date"].dt.month
df["day"] = df["date"].dt.day
print(df[["date",
          "month" , "day"]].head())
```

```python
df.dtypes
```

```python
# Count missing values
print(df.isnull().sum())
```

```python
missing_pct = (df.isnull().sum()/len(df)*100).round(1)
print(missing_pct)
```

```python
# filling missing numbers with the column average
df["total _ amount"]= df["total _ amount"].fillna(df["total _ amount"].mean())
df["rating"]= df["rating"].fillna(df["rating"].mean())
print("missing in rating:",df["rating"].isnull().sum())
print("missing in total:" , df["total _ amount"].isnull().sum())
```

```python
df["pharmacist _ name"] = df["pharmacist _ name"].fillna("unknown")
```

```python
# check how many rows still have missing values
print("Remaining missing values:")
print(df.isnull().sum())
```

```python
# Drop the remaining rows with misssing values
print("Row before dropping:", len(df))
df= df.dropna()
print("Rows after droping:" , len(df))
```

```python
print("missing value remaining:" )
print(df.isnull().sum())
print("Final Shape:" , df.shape)
```
