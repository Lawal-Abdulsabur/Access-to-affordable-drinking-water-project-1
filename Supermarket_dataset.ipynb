# Cleaning of the dataset

```python
# Fix date datatype and Extract month and day of week
df["Date"] = pd.to_datetime(df["Date"])

df["month"] = df["Date"].dt.month_name()

df["day_of_week"] = df["Date"].dt.day_name()

df[[ "Date" ,"month", "day_of_week"]]
```

# I fixed the date column and extracted the days and months from the date

```python
# # Step 1 — clean all column names at once
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')


print( "Cleaned columns name")
df.columns
```

```python
# Product line the generated the most revenue
Product_line_revenue =df.groupby("product_line")["total"].sum().sort_values( ascending = False)

Product_line_revenue
```

```python
# City with highest customer rating

City_customer_rating = (
    df.groupby("city")["rating"]
      .agg(["count", "max"])
      .sort_values("max", ascending=False)
)

City_customer_rating
```

# The city with the highest customer rating is Mandalay with 332 counts

```python
# Average spend — Member vs Normal customers

Average_total_spending = df.groupby("customer_type")["total"].mean().sort_values( ascending= False)
Average_total_spending

```

# Customer_type Member has an average of 327.791305 spendings

# Visualization  
# Which product line generate the most revenue

```python
# Bar chart: Total revenue by product line

revenue_by_product = df.groupby("product_line")["total"].sum().sort_values(ascending=False)

# Step 2 — create the chart
plt.figure(figsize=(10, 5))
plt.bar(revenue_by_product.index, revenue_by_product.values, color="steelblue")

# Step 3 — add labels and title
plt.title("Total Revenue by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

# Bar chart

# Which branch is performing best

```python
# Aggregate total revenue per branch
branch_totals = df.groupby("branch")["total"].sum()

# Plot bar chart
plt.figure(figsize=(10, 5))
plt.bar(branch_totals.index, branch_totals.values, color="orange", edgecolor="white")

plt.title("Total Revenue per Branch")
plt.xlabel("Branch")
plt.ylabel("Total Revenue")
plt.show()
```

# Line Chart

```python
City_customer_rating_sorted = City_customer_rating.sort_values("max")

plt.figure(figsize=(10, 5))
plt.plot(City_customer_rating_sorted.index, City_customer_rating_sorted["max"],
         marker="o", color="steelblue", linewidth=2, markersize=6)

plt.title("Highest Customer Rating by City")
plt.xlabel("City")
plt.ylabel("Max Rating")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
